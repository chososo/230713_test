```python
import requests
import os

def download_images(query, count):
    # API 요청에 필요한 정보들
    client_id = 'iV1Cx31Pzm5VXF703AK7'  # 네이버 개발자 센터에서 발급받은 클라이언트 ID
    client_secret = '8WS9CiT8nt'  # 네이버 개발자 센터에서 발급받은 클라이언트 시크릿

    # API 요청 URL
    url = 'https://openapi.naver.com/v1/search/image'
    headers = {'X-Naver-Client-Id': client_id, 'X-Naver-Client-Secret': client_secret}

    # API 요청 파라미터 설정
    params = {'query': query, 'display': count}

    # API 요청 보내기
    response = requests.get(url, headers=headers, params=params)
    data = response.json()

    # 이미지 저장 디렉토리 생성
    os.makedirs(query, exist_ok=True)

    # 이미지 다운로드
    for i, item in enumerate(data['items']):
        image_url = item['link']
        image_name = f"{query}_{i+1}.jpg"  # 이미지 파일명 설정
        save_path = os.path.join(query, image_name)  # 이미지 저장 경로 설정

        # 이미지 다운로드 요청
        image_response = requests.get(image_url)
        with open(save_path, 'wb') as f:
            f.write(image_response.content)

        print(f"다운로드 완료: {save_path}")

# 이미지 다운로드 함수 호출
download_images('뉴진스 혜린', 60)  # '뉴진스 민지'라는 검색어로 60개의 이미지를 다운로드
```
